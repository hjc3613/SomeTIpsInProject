```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(org_func, ft, sampling_period=3):
    n = len(org_func)
    interval = sampling_period / n
    # 原始函数
    plt.subplot(2,1,1)
    plt.plot(np.arange(0, sampling_period, interval), org_func, 'black')
    plt.xlabel('time')
    plt.ylabel('amplitude')
    # 变换后的函数
    plt.subplot(2,1,2)
    plt.plot()
    frequency = np.arange(n/2)/(n*interval)
    nfft = abs(ft[range(int(n/2))] / n)
    plt.plot(frequency, nfft, 'red')
    plt.xlabel('freq/hz'),
    plt.ylabel('amp.spectrum')
    plt.show()
```

```python
time = np.arange(0, 5, 0.05)
x = np.sin(2 * np.pi * 1 * time)
y = np.fft.fft(x)
```

```python
show(x, y)
```

```python
from scipy.io import wavfile
datasound = '/home/hujunchao/Videos/test.wav'
fs, data = wavfile.read(datasound)
data = np.array(data, dtype=float)
data = (data-np.mean(data))/np.std(data)
time = range(len(data))
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.plot(time, data)
plt.show()
```

```python
print(len(data), np.shape(data), fs)
```

```python
fftdata = np.fft.fft(data)
```

```python
fftdata[:10]
```

```python
fftdatafreq = np.zeros((len(data)))
for i in range(len(fftdata)):
 fftdatafreq[i] = abs(fftdata[i].real)
```

```python
fftdatafreq[:10]
```

```python
plt.ylabel("Amplitude")
plt.xlabel("Frequency")
plt.plot(time, fftdatafreq)

plt.show()
```
